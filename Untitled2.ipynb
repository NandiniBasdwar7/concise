```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/customers.csv')
df2 = pd.read_csv('/content/orders.csv')
df3 = pd.read_csv('/content/products.csv')
```

```python
df_merged = pd.merge(df1, df2)
```

```python
all_merged = pd.merge(df_merged, df3)
```

```python
all_merged.head()
```

```python
all_merged['revenue'] = all_merged['quantity'] * all_merged['price']
```

```python
all_merged['order_date']= pd.to_datetime(all_merged['order_date'])
```

```python
all_merged['month']= all_merged['order_date'].dt.month
```

```python
all_merged['revenue'] = all_merged['quantity'] * all_merged['price']
all_merged.groupby('month')['revenue'].sum()
```

```python
purchases = all_merged.groupby(['country','customer_name'])['revenue'].sum().reset_index()
```
