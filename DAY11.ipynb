```python
#Groupby():You typically call groupby() on a DataFrame and pass the column name(s) you want to group by:
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A'],
        'Value': [10, 15, 12, 8, 20, 11]}
df = pd.DataFrame(data)

# Group by a single column
grouped_df = df.groupby('Category')

# Group by multiple columns (pass a list of column names)
grouped_df_multi = df.groupby(['Category', 'Value'])

```

```python
#Aggregation:

#.sum(): Calculates the sum of values in each group.
# .mean(): Calculates the mean of values in each group.
# .count(): Counts the number of non-null values in each group.
# .min()/.max(): Finds the minimum/maximum value in each group.
# .agg(): Allows applying multiple aggregation functions simultaneously.
category_sums = df.groupby('Category')['Value'].sum()
print(category_sums)
```

```python
category_mean= df.groupby('Category')['Value'].mean()
category_mean
```

```python
category_count=df.groupby('Category')['Value'].count()
category_count
```

```python
#pandas.pivot_table():
# index:
# Specifies the column(s) to use as the new row index of the pivot table. This creates the "rows" of your summarized data.
# columns:
# Specifies the column(s) to use as the new column headers of the pivot table. This creates the "columns" of your summarized data.
# values:
# Specifies the column(s) whose values will be aggregated and displayed in the pivot table's cells.
# aggfunc:
# Defines the aggregation function(s) to apply to the values column(s). Common aggregation functions include 'mean', 'sum', 'count', 'min', 'max', etc. You can also provide a list of functions or a dictionary to apply different functions to different value columns.
import pandas as pd

data = {
    'Region': ['East', 'West', 'East', 'West', 'East'],
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Sales': [100, 150, 120, 200, 180]
}
df = pd.DataFrame(data)

# Create a pivot table to show total sales by Region and Product
pivot_df = df.pivot_table(index='Region', columns='Product', values='Sales', aggfunc='sum')
print(pivot_df)
```

```python
#Excercise
import pandas as pd
import numpy as np

# Sample sales data
data = {
    'Date': pd.to_datetime(['2025-07-01', '2025-07-01', '2025-07-02', '2025-07-08', '2025-07-15', '2025-07-15', '2025-07-20', '2025-07-22', '2025-08-01', '2025-08-05']),
    'Sales': [100, 150, 200, 300, 250, 350, 400, 450, 500, 550]
}
df_sales = pd.DataFrame(data)
df_sales
df_sales['Date'] = pd.to_datetime(df_sales['Date'])
daily_sales_pivot = pd.pivot_table(df_sales,
                                   values='Sales',
                                   index=pd.Grouper(key='Date', freq='D'),
                                   aggfunc=np.sum)

print(daily_sales_pivot)
weekly_sales_pivot = pd.pivot_table(df_sales,
                                    values='Sales',
                                    index=pd.Grouper(key='Date', freq='W'),
                                    aggfunc=np.sum)

print( weekly_sales_pivot)


monthly_sales_pivot = pd.pivot_table(df_sales,
                                     values='Sales',
                                     index='Date',
                                     aggfunc=np.sum)
print(monthly_sales_pivot)
```

```python
#
import pandas as pd

# Sample sales data with Region and Category
data = {
    'Region': ['East', 'West', 'East', 'West', 'East', 'North', 'South', 'West'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing'],
    'Sales': [100, 200, 150, 300, 250, 180, 220, 280]
}
df = pd.DataFrame(data)
df
```

```python
# Use rolling averages to smooth out daily sales data over 7-day windows.
import pandas as pd

# Sample daily sales data with a DatetimeIndex
dates = pd.date_range(start='2025-07-01', periods=15)
sales_data = [
    120, 130, 110, 140, 125, 150, 160, # First 7 days
    170, 180, 165, 190, 175, 200, 210, # Next 7 days
    220  # Day 15
]
df_daily_sales = pd.DataFrame({'Sales': sales_data}, index=dates)

df_daily_sales
```

```python
df_daily_sales['7_Day_Rolling_Avg_Sales']=df_daily_sales['Sales'].rolling(window=7,min_periods=1).mean()
df_daily_sales
```

```python
#Join()
import pandas as pd

df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]}, index=['A', 'B'])
df2 = pd.DataFrame({'City': ['New York', 'London'], 'Occupation': ['Engineer', 'Artist']}, index=['A', 'C'])

# Left join based on index
result = df1.join(df2,how='outer')
print(result)
```

```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']}, index=[0, 1])
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']}, index=[2, 3])

# Concatenate row-wise (default)
result_rows = pd.concat([df1, df2])
print("Concatenated Rows:")
print(result_rows)

df3 = pd.DataFrame({'C': ['C0', 'C1']}, index=[0, 1])

# Concatenate column-wise
result_cols = pd.concat([df1, df3])
print("\nConcatenated Columns:")
print(result_cols)
```
