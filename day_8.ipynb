```python
#Creating a Series-A Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, Python objects, etc.). The labels are collectively called the index,
import pandas as pd
my_list = [10, 20, 30, 40, 50]
my_series = pd.Series(my_list)
print(my_series)
```

```python
#Creating a DataFrame-A Pandas DataFrame is a two-dimensional, labeled data structure with columns of potentially different types, similar to a spreadsheet or SQL table. You can create a DataFrame in numerous ways: From a Dictionary of Lists (Common Method): This is a very common and intuitive way to create DataFrames where keys become column names and values (lists) represent the data in those columns.
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'London', 'Paris']}
df = pd.DataFrame(data)
print(df)
```

```python
#From a List of Lists: Each inner list represents a row in the DataFrame. You can also specify column names and index labels.
import pandas as pd
data = [['Alice', 25, 'New York'],
        ['Bob', 30, 'London'],
        ['Charlie', 35, 'Paris']]
df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
print(df)
```

```python
#From a Dictionary of Series: You can combine multiple Series objects into a DataFrame, where each Series becomes a column,
import pandas as pd
s1 = pd.Series([10, 20, 30], name='Numbers')
s2 = pd.Series(['apple', 'banana', 'orange'], name='Fruits')
df = pd.concat([s1, s2], axis=1) # Concatenate along columns (axis=1)
print(df)
```

```python
#With Index Labels: You can provide custom row index labels when creating a DataFrame.
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data, index=['student_a', 'student_b', 'student_c'])
print(df)
```

```python
# Exercise 11:-Load a sample CSV (e.g., sales.csv).
#Print column types, shape, top 5 rows.
#Rename one column.
#
import pandas as pd

try:
    df = pd.read_csv('/content/sales_data_sample.csv', encoding='latin1')
    print("File read successfully with latin1 encoding.")
except UnicodeDecodeError:
    try:
        df = pd.read_csv('/content/sales_data_sample.csv', encoding='cp1252')
        print("File read successfully with cp1252 encoding.")
    except UnicodeDecodeError:
        try:
            df = pd.read_csv('/content/sales_data_sample.csv', encoding='utf-8-sig') # For files with a BOM
            print("File read successfully with utf-8-sig encoding.")
        except UnicodeDecodeError:
            print("Could not decode with common encodings. Consider trying to detect the encoding or re-saving the file.")
print(df.head(10))

```

```python
print("Column Types:")
print(df.dtypes)  # dtypes attribute returns the data type of each column.
print("\\n")

# 2. Print DataFrame Shape
print("DataFrame Shape:")
print(df.shape)  # shape attribute returns a tuple (rows, columns).
print("\\n")

# 3. Print Top 5 Rows
print("Top 5 Rows:")
print(df.head(5))
```

```python
df.rename(columns={'ORDERNUMBER': 'Order_ID'}, inplace=True)

# Print the columns to verify the change
print(df.columns)
print("\\nTop 5 Rows after renaming:")
print(df.head())
```

```python
columns_to_drop = ['ADDRESSLINE2', 'TERRITORY']

# Drop the specified columns
# axis=1 indicates that columns are dropped
# inplace=True modifies the DataFrame directly
df.drop(columns=columns_to_drop, axis=1, inplace=True)

# Verify the columns are dropped by printing the DataFrame's columns
print("DataFrame columns after dropping:")
print(df.columns)
```

```python

# Assuming df is your loaded DataFrame
# df = pd.read_csv('/content/sales_data_sample.csv', encoding='latin1')
# And you've already dropped ADDRESSLINE2 and TERRITORY

# Calculate the mode of the 'STATE' column
state_mode = df['STATE'].mode()[0]  # .mode() returns a Series, so [0] gets the first mode

# Fill missing values in 'STATE' with the calculated mode
df['STATE'].fillna(state_mode, inplace=True)

print("Missing values in 'STATE' after mode imputation:")
print(df['STATE'].isnull().sum())
print("\\nFirst 5 rows of 'STATE' column after imputation:")
print(df['STATE'].head())
```

```python
import pandas as pd

# Calculate the mode (most frequent value) of the 'POSTALCODE' column
mode_postalcode = df['POSTALCODE'].mode()[0]

# Fill the missing values in the 'POSTALCODE' column with the calculated mode
df['POSTALCODE'].fillna(value=mode_postalcode, inplace=True)
print(df['POSTALCODE'].isnull().sum())
print(df['POSTALCODE'].head())
```
