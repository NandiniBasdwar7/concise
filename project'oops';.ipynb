```python
# Project Overview:Build a modular pipeline using OOP for processing data files and logging results.
#data loader class
import pandas as pd
class DataLoader:
  def __init__(self,filepath):
    self.filepath=filepath
  def load_data(self):
    if self.filepath.endswith('.csv'):
      return pd.read_csv(self.filepath)
    else:
      raise ValueError('Unsupported file format')
    return
obj=DataLoader('/content/index.csv')
obj.load_data()
```

```python
#datacleaner class
class DataCleaner:
  def __init__(self,data):
    self.data=data
  def clean_data(self):
    self.data.fillna('unknown',inplace=True)
    self.data.dropna(inplace=True)
    self.data.dropna(inplace=True)
    return self.data
obj=DataLoader("/content/index.csv")
obj=obj.load_data()
obj=DataCleaner(obj)
obj.clean_data()
```

```python
#class data transformer- creat new coloumn 'high_value'.
class DataTransformer:
  def __init__(self,data):
    self.data=data
  def transform_data(self):
    self.data['high_value']=self.data['money'].apply(lambda x:1 if x>30 else 0)
    return self.data
obj=DataLoader("/content/index.csv")
obj=obj.load_data()
obj=DataTransformer(obj)
obj.transform_data()
```

```python
#Logger – logs events with timestamps.
import logging
class Logger:
  def __init__(self,filename):
    self.filename=filename
  def log_event(self,event):
    logging.basicConfig(filename=self.filename,level=logging.INFO,format='%(asctime)s %(message)s')
    logging.info(event)
```
